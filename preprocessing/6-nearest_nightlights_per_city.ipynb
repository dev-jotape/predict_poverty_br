```python
import pandas as pd
```

```python
df = pd.read_csv('../model/coordinates_with_distances.csv')
df
```

```python
# remove < 0
df = df[df['radiance'] > 0]
df
```

```python
df['rank'] = df.groupby('city_code')['distance_to_city'].rank()
nearest = df[df['rank'] <= 100]
nearest
```

```python
nearest.to_csv('../model/nearest_nightlights_per_city.csv', index=False)
```
